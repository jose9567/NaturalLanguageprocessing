```python
import string
def remove_punctuation(text):
  punctuationfree=' '
  punctuationfree="".join([i for i in text if i not in string.punctuation])
  return punctuationfree

data['clean_tweet']=data['tweet'].apply(lambda x:remove_punctuation(x))
  #to remove punctuation
```

```python
data.head()
```

```python
string.punctuation
```

```python
#word tokenization function
import nltk
nltk.download('punkt')
def tokenization(text):
  tokens=nltk.word_tokenize(text)
  return tokens

data['tokenized_tweet']=data['clean_tweet'].apply(lambda x:tokenization(x))
```

```python
data.head()
```

```python
#function to remove stopwords
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  output= [i for i in text if i not in stopwords]
  return output

data['remove_stopwords']=data['tokenized_tweet'].apply(lambda x:remove_stopwords(x))
```

```python
data.head()
```

```python
#function for lemmatization
from nltk.stem import WordNetLemmatizer # Capitalize 'W' in 'WordNetLemmatizer'
nltk.download('wordnet')
wordnet_lemmatizer=WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text

data['lemmatized_tweet']=data['remove_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
data.head()
```

```python
corpus = [' '.join(tokens) for tokens in data['lemmatized_tweet']]
```

```python
corpus[0:10]
```

```python
#Bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
x=cv.fit_transform(corpus)
y=data.label
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
from sklearn import metrics
y_pred=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_pred,y_test))
```

Transform the input text, rename variable to avoid shadowing built-in 'input'

```python
def find_sentiment(text):
  input_text = cv.transform([text])
  pred = clf.predict(input_text)[0]
  if pred == 1:
    return "Negative"
  elif pred == 0:
    return "Positive"
```

```python
input= 'What amazing service Apple wont even talk question I unless I pay 1995 stupid support'
find_sentiment(input)
```

TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def vectorizer_tfidf(corpus):
  tfidf_model=TfidfVectorizer()
  tfidf_matrix=tfidf_model.fit_transform(corpus).todense()
  tfidf_df=pd.DataFrame(tfidf_matrix)
  tfidf_df.columns=sorted(tfidf_model.vocabulary_)
  return tfidf_df
```
